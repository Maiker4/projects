```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = "diamonds.csv"
```

```python
diamonds_raw = pd.read_csv(data)

diamonds_raw.head()
```

```python
def rename_columns(old):
    new_names = {'x':'lengh','y':'widht','z':'depth', 'carat': 'weight'}
    new = old.rename(columns=new_names)

    return new
```

```python
diam_new = rename_columns(diamonds_raw).drop(columns='Unnamed: 0')
diam_new.head()
```

```python
# diam_new.value_counts
print(diam_new['cut'].drop_duplicates().count())
print(diam_new['cut'].drop_duplicates())
print('')
print(diam_new['color'].drop_duplicates().count())
print(diam_new['color'].drop_duplicates())
print('')
print(diam_new['clarity'].drop_duplicates().count())
print(diam_new['clarity'].drop_duplicates())

```

```python
cut_quality = {'Fair': 1,
               'Good': 2,
               'Very Good': 3,
               'Premium': 4,
               'Ideal': 5
               }

color_quality = {'J': 1,
                 'I': 2,
                 'H': 3,
                 'G': 4,
                 'F': 5,
                 'E': 6,
                 'D': 7
                 }

clarity_quality = {'I1': 1,
                   'SI2': 2,
                   'SI1': 3,
                   'VS2': 4,
                   'VS1': 5,
                   'VVS2': 6,
                   'VVS1': 7,
                   'IF' : 8
                   }
```

```python
diam_new['cut_encoded'] = diam_new['cut'].map(cut_quality)
diam_new['color_encoded'] = diam_new['color'].map(color_quality)
diam_new['clarity_encoded'] = diam_new['clarity'].map(clarity_quality)

# diam_new.head()

diam_corr_matrix = diam_new.drop(columns=['cut', 'color', 'clarity'])

diam_corr_matrix.dtypes
```

```python
matrix = diam_corr_matrix.corr()
matrix
```
